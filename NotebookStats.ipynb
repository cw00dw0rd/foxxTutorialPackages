```python
# importing the requests library 
import requests 

# api-endpoint 
URL = "https://d383fa0b596a.arangodb.cloud:8529/_db/_system/stats/tutorialStats"

# AISIS tutorial endpoint
#URL = "https://a0434a558688.arangodb.cloud:8529/_db/_system/stats/tutorialStats"

# AISIS ML endpoint
#URL = "https://a0434a558688.arangodb.cloud:8529/_db/_system/stats/aisisStats"

# sending get request and saving the response as response object 
r = requests.get(url = URL) 

# extracting data in json format 
data = r.json() 
```

```python
expired = data['stats']['_documents'][0][0]
active = data['stats']['_documents'][0][1]
all = expired + active
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(all)
```

```python
agg = df['TutorialName'].value_counts()
print(agg)
agg.plot(kind='bar')
```
